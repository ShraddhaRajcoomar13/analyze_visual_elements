```python
from PIL import Image
import cv2
import numpy as np
import os
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
import pandas as pd
import re
```

```python
def space(img):
    
    # Load the image
    image = cv2.imread(img, cv2.IMREAD_GRAYSCALE)
    image = cv2.cvtColor(color_image, cv2.COLOR_BGR2GRAY)
    # Calculate space ratio
    total_pixels = image.size
    white_pixels = np.sum(image == 255)  # Assuming white is represented as 255
    space_ratio = white_pixels / total_pixels

    return space_ratio
```

```python
def chaos(img):
    # Load the image
    image = cv2.imread(img, cv2.IMREAD_GRAYSCALE)
    image = cv2.cvtColor(color_image, cv2.COLOR_BGR2GRAY)
    # Calculate chaos as the standard deviation of pixel values
    chaos = np.std(image)
    return chaos
```

```python
def stab(img):
    # Load the image
    image = cv2.imread(img, cv2.IMREAD_GRAYSCALE)
    image = cv2.cvtColor(color_image, cv2.COLOR_BGR2GRAY)
    # Apply Canny edge detection
    edges = cv2.Canny(image, threshold1=30, threshold2=100)  # Adjust thresholds as needed
    # Calculate the line stability by counting detected lines
    lines = cv2.HoughLinesP(edges, 1, np.pi / 180, threshold=50, minLineLength=50, maxLineGap=10)
    line_stability = len(lines)
    return line_stability
```

```python
def add_space_before_capitals(input_string):
    # Use regular expression to find all capital letters
    result = re.sub(r'([A-Z])', r' \1', input_string)
    return result
```

```python
p={}

# Define the directory path
directory_path = r"C:\Users\shraDDhar\OneDrive - TEVO (Pty) Ltd\Desktop\drawings\piggy"

chaosm=[]
stabs=[]
spaces=[]
compound_scores=[]
compound_score_labels=[]
descriptions=[]

# List files in the directory
files = os.listdir(directory_path)

num=0

# Print the list of files
for file in files:
    num+=1
    # Read a color image
    color_image = cv2.imread(directory_path+'/'+file)
    # Convert the color image to grayscale
    grayscale_image = cv2.cvtColor(color_image, cv2.COLOR_BGR2GRAY)
    # Create a SentimentIntensityAnalyzer object
    analyzer = SentimentIntensityAnalyzer()

    # Get sentiment scores
    sentiment_scores = analyzer.polarity_scores(add_space_before_capitals(file[:-4]))

    # Extract the compound score
    compound_score = sentiment_scores['compound']

    # Interpret the compound score
    if compound_score >= 0.05:
        sentiment = "Positive"
    elif compound_score <= -0.05:
        sentiment = "Negative"
    else:
        sentiment = "Neutral"
    print(num)
    chaosm.append(chaos(directory_path+'/'+file))
    stabs.append(stab(directory_path+'/'+file))
    spaces.append(space(directory_path+'/'+file))
    compound_scores.append(compound_score)
    compound_score_labels.append(sentiment)
    descriptions.append(add_space_before_capitals(file[:-4]))
    #new_file_name = 'batch_1_'+str(num)+'.jpg'
    ######################################################### Rename the image file
    
p["compound_score"]= compound_scores
p['compound_score_label']=compound_score_labels
p['description']=descriptions
p['chaos']=chaosm
p['space_ratio']=spaces
p['line_stability']=stabs

pdf=pd.DataFrame(p)
```
